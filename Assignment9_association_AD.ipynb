# Book data

```python
import pandas as pd
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
book = pd.read_csv("book.csv")
book
```

```python
for i in book.columns:
    print(i) 
    print(book[i].value_counts())
    print()
    
```

```python
#making rule with support of 0.1 and confidence of 0.6
book_freitemset1 = apriori(book,min_support=0.1,use_colnames=True)
book_freitemset1
```

```python
bookrule1= association_rules(book_freitemset1,metric='confidence',min_threshold=0.6)
bookrule1
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
sn.pairplot(bookrule1)
```

```python
#making rule with support of 0.15 and confidence of 0.6
book_freitemset2 = apriori(book,min_support=0.15,use_colnames=True)
book_freitemset2
```

```python
bookrule2= association_rules(book_freitemset2,metric='confidence',min_threshold=0.6)
bookrule2
```

```python
sn.pairplot(bookrule2)
```

# Movies data

```python
movies = pd.read_csv("my_movies.csv")
movies
```

```python
for i in movies.columns:
    print(i) 
    print(movies[i].value_counts())
    print()
```

```python
movies= pd.get_dummies(movies)
movies
```
